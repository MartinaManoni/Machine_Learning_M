#####     **K-Folds cross-validator**
##### Provides train/test indices to split data in train/test sets. Split dataset into k consecutive folds (without shuffling by default).
##### Each fold is then used once as a validation while the k - 1 remaining folds form the training set.

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=10, shuffle=True)
kf.get_n_splits(df['target']) #we work only with the target feature
print(kf)


for train_index, test_index in kf.split(df['target']):
  print("TRAIN:", train_index, "VALIDATION:", test_index)
  X_train, X_test = df['target'][train_index], df['target'][test_index]
  y_train, y_test = df['target'][train_index], df['target'][test_index]
```

In k fold cross validation we have no overlap between validation folds as we can see below

```python
for train_index, test_index in kf.split(df['target']):
  print("VALIDATION:", test_index)
 
```

##### Repeats K-Fold n times with different randomization in each repetition

```python
from sklearn.model_selection import RepeatedKFold
#random_state=2652124
rkf = RepeatedKFold(n_splits=10, n_repeats=2, random_state=2652124)
for train_index, test_index in rkf.split(df['target']):
  print("TRAIN:", train_index, "VALIDATION:", test_index)
  X_train, X_test = df['target'][train_index], df['target'][test_index]
  y_train, y_test = df['target'][train_index], df['target'][test_index]
```

# K FOLD & SVC (TOTAL IRIS DATASET)

```python
# Importing the necessary Python libraries
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split, KFold
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import svm
```

```python
iris = datasets.load_iris()
#iris.head()
X_features = pd.DataFrame(data = iris['data'], columns = iris['feature_names'])
y_target = pd.DataFrame(data = iris['target'], columns = ['target'])

kf = KFold(n_splits=10, shuffle=True)
print(kf)
#X_features.head()
```

```python
import statistics
accuracy_vec = []

# Iterating through each of the folds in K-Fold
for train_index, val_index in kf.split(X_features):

# Splitting the training set from the validation set for this specific fold
    X_train, X_val = X_features.iloc[train_index, :], X_features.iloc[val_index, :]
    y_train, y_val = y_target.iloc[train_index], y_target.iloc[val_index]
    
    # Fitting the X_train and y_train datasets to the SVC model
    clf = svm.SVC(kernel='rbf', C=1).fit(X_train,y_train.values.ravel())
    #print('accuracy:', clf.score(X_val,y_val))
    accuracy_vec.append(clf.score(X_val,y_val.values.ravel()))
    
    # Getting inferential predictions for the validation dataset
    #val_preds = rfc_model.predict(X_val)
    
    # Generating validation metrics by comparing the inferential predictions (val_preds) to the actuals (y_val)
    #val_accuracy = accuracy_score(y_val, val_preds)
    #val_confusion_matrix = confusion_matrix(y_val, val_preds)
    
    # Printing out the validation metrics
    #print(f'Accuracy Score: {val_accuracy}')
    #print(f'Confusion Matrix: \n{val_confusion_matrix}')

print('accuracy vector', accuracy_vec)
#print(X_train.shape)
#print(y_train.shape)
#print(y_train.values.ravel().shape)
x = statistics.mean(accuracy_vec)
print('MEAN ACCURACY', x)
```

# HOLD OUT VALIDATION SCHEMA & SVC



```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm

X, y = datasets.load_iris(return_X_y=True)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
#print(X_train.shape, y_train.shape)
#print(X_test.shape, y_test.shape)

clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
print('accuracy:', clf.score(X_test, y_test))
```
